```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("datasets/amazon_jobs_dataset.csv",skipinitialspace =True)
df

```

```python
date=np.array(df.Posting_date)
yr=[]
for item in date:
    
    x=item.split(",")
    yr.append(x[1])
    
print(yr)
```

```python
df.Title.isnull().sum()
```

```python
# means every posting date gives job 
```

```python
new_yr=np.array(yr,dtype=int)

Y=[]
for i in range(2011,2019):
    
    
    a=np.count_nonzero(new_yr == i)
    Y.append(a)
print(Y)
```

```python
X=np.arange(2011,2019)
```

# PLOTING

```python
plt.plot(X,Y,linewidth=5,color='red')
plt.xlabel("YEAR")
plt.ylabel("JOB")
plt.grid()

plt.show()
```
